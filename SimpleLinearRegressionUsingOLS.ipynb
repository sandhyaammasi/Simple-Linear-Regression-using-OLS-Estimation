## Import the relevant libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import statsmodels.api as sm 
sns.set() #resets the matplot lib into seaborn style graphs automatically
```

## Load the data

```python
df = pd.read_csv("real_estate_price_size.csv")
df.head()
```

```python
df.describe()
```

## Create the regression

### Declare the dependent and the independent variables

```python
 df.columns
 x1 = df["size"]
 y = df["price"]
 df.isnull().sum()
```

### Explore the data

```python
plt.scatter(x1,y)
plt.xlabel("SIZE",fontsize = 20)
plt.ylabel("PRICE",fontsize = 20)
plt.show
```

### Regression itself

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

### Plot the regression line on the initial scatter

```python
plt.scatter(x1,y)
## yhat = x1 * slope + intercept(co eff)
yhat = 223.1787	* x1 + 101900
plt.plot(x1,yhat, c= "GREEN",lw = 6)
plt.xlabel('SIZE')
plt.ylabel('PRICE')
```
